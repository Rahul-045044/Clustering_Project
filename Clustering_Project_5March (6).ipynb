###Liveness:

Liveness measures the presence of a live audience in the recording, with values ranging from 0 to 1. A higher value indicates a higher probability of a live performance.

###Loudness:

Loudness denotes the overall loudness of the track in decibels (dB), providing information about its volume and intensity.

###Popularity:

Popularity indicates the popularity of the track, reflecting its reception and engagement among listeners.

###Speechiness:

Speechiness measures the presence of spoken words in the track, ranging from 0 to 1. Higher values indicate tracks with more spoken-word elements, such as podcasts or spoken-word performances.

###Tempo:

Tempo represents the speed or pace of the track in beats per minute (BPM), providing information about its rhythmic structure and tempo.

```python
wcssd = [] # Within-Cluster-Sum-Squared-Distance
nr_clus = range(1,11) # Number of Clusters
for k in nr_clus:
    kmeans = kmclus(n_clusters=k, init='random', random_state=111)
    kmeans.fit(df_music)
    wcssd.append(kmeans.inertia_)
plt.plot(nr_clus, wcssd, marker='x')
plt.xlabel('Values of K')
plt.ylabel('Within Cluster Sum Squared Distance')
plt.title('Elbow Curve for Optimal K')
plt.show()
```

```python
km_2cluster = kmclus(n_clusters=2, init='random', random_state=222)
km_2cluster_model = km_2cluster.fit_predict(df_music); km_2cluster_model
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler

# Assuming you have a DataFrame called 'df_ppd' with preprocessed data
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_gauss)  # Corrected line

# Initialize and fit the Gaussian Mixture Model
n_components = 2 # You can adjust the number of components/clusters
gmm = GaussianMixture(n_components=n_components, random_state=142)
gmm.fit(X_scaled)

# Get cluster labels
cluster_labels = gmm.predict(X_scaled)

# Add cluster labels to the DataFrame
df_gauss['Cluster'] = cluster_labels

# Display the DataFrame with cluster labels
df_gauss
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Calculate silhouette score
silhouette_score_value = silhouette_score(X_scaled, cluster_labels)

# Calculate Davies-Bouldin score
davies_bouldin_score_value = davies_bouldin_score(X_scaled, cluster_labels)

print("Silhouette Score:", silhouette_score_value)
print("Davies-Bouldin Score:", davies_bouldin_score_value)
```
